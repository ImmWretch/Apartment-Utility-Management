```python
ds = ds.drop('water_use_l',axis = 1)
```

```python
ds = ds.drop('water_use_l_high_res', axis = 1)
ds = ds.drop('washing', axis=1)
ds = ds.drop('dishwasher', axis =1 )
```

```python
ds.head()
```

```python
ds.info()
```

```python
ds["date_time"] = pd.to_datetime(ds["date_time"])
```

```python
ds["hour"] = ds["date_time"].dt.hour
ds["day_of_week"] = ds["date_time"].dt.dayofweek
```

```python
ds
```

```python

# Load the dataset


# Convert date_time to datetime (if not already)
ds["date_time"] = pd.to_datetime(ds["date_time"], errors="coerce")

# Drop rows with missing target values
ds = ds.dropna(subset=["electrcity_use_kwh"])

# Drop the raw datetime column (features already extracted)
ds = ds.drop(columns=["date_time"])

# Convert boolean to integer
ds["occupied"] = ds["occupied"].astype(int)

# One-hot encode the building column
ds_encoded = pd.get_dummies(
    ds,
    columns=["building"],
    prefix="building"
)

# Save the processed dataset
ds_encoded.to_csv("electricity.csv", index=False)

print("One-hot encoding complete. Saved as electricity.csv")
```
